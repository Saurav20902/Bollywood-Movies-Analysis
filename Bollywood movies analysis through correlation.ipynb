```python

import pandas as pd
import numpy as np
import seaborn as sns
# Importing all the libraries and packages I need for this project
import matplotlib.pyplot as plt # import matplot from matlab
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline                           #we use this to render the figure   
matplotlib.rcParams['figure.figsize'] = (12,8)  #plot configurations




# Now we need to read in the data
df = pd.read_csv(r'C:\Users\saura\Downloads\movies.csv') # here we put the file into play to conduct an analysis of it
```

```python
df.head() # to get a peek at the data
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Order our Data a little bit to see

df.sort_values(by=['Revenue(INR)'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.drop_duplicates()
```

```python
plt.scatter(x=df['Budget(INR)'], y=df['Revenue(INR)'], alpha=0.5) #setting x-axis and y-axis
plt.title('Budget vs Revenue Earnings') #setting the title
plt.xlabel('Revenue Earnings') #setting the labels
plt.ylabel('Budget for Film')
plt.show() #generating the graph
```

```python
sns.regplot(x="Revenue(INR)", y="Budget(INR)", data=df) #plotting correlation plot on the basis of positive slope
```

```python
df.corr() # checking correlation
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
CompanyGrossSum = df.groupby('Lead Star')[["Revenue(INR)"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('Revenue(INR)', ascending = False)[:5]

CompanyGrossSumSorted = CompanyGrossSumSorted['Revenue(INR)'].astype('int64') 

CompanyGrossSumSorted
```
